# Pemformatan/Perapian Data

```python
# Mengubah nama kolom menggunakan regex
columns_update = {
    "Place_Id": "Id_Wisata",
    "Place_Name": "Nama_Wisata",
    "Description": "Deskripsi",
    "Category": "Kategori",
    "City": "Kota",
    "Price": "Harga",
    "Rating": "Rating",
    "Time_Minutes": "Waktu per Menit",
    "Coordinate": "Koordinat",
    "Lat": "Latitude",
    "Long": "Longitude",
    "Unnamed: 11": "Tanpa Nama 1",
    "Unnamed: 12": "Tanpa Nama 2"
}

for old_columns, new_columns in columns_update.items():
    df.columns = df.columns.str.replace(old_columns, new_columns, regex=True)
    
df.head()
```

```python
# Mengubah posisi kolom
df = df[[
    "Id_Wisata",
    "Nama_Wisata",
    "Deskripsi",
    "Kategori",
    "Kota",
    "Harga",
    "Waktu per Menit",
    "Rating",
    "Koordinat",
    "Latitude",
    "Longitude",
    "Tanpa Nama 1",
    "Tanpa Nama 2"
]]

df.head()
```

```python
# Menjumlahkan kolom harga yang tidak memiliki nilai 0 berdasarkan rating yang tertinggi di Yogyakarta 
jumlah_bayar_yogyakarta = sum(yogyakarta_rating["Harga"] != 0)
jumlah_bayar_yogyakarta
```

```python
# Menjumlahkan kolom harga yang tidak memiliki nilai 0 berdasarkan rating yang tertinggi di Bandung 
jumlah_bayar_bandung = sum(bandung_rating["Harga"] != 0)
jumlah_bayar_bandung
```

```python
# Menjumlahkan kolom harga yang tidak memiliki nilai 0 berdasarkan rating yang tertinggi di Semarang 
jumlah_bayar_semarang = sum(semarang_rating["Harga"] != 0)
jumlah_bayar_semarang
```

```python
# Menjumlahkan kolom harga yang tidak memiliki nilai 0 berdasarkan rating yang tertinggi di Surabaya 
jumlah_bayar_surabaya = sum(surabaya_rating["Harga"] != 0)
jumlah_bayar_surabaya
```

```python
# Menjumlahkan semua kolom harga yang tidak memiliki nilai 0 berdasarkan rating yang tertinggi di setiap kota
jumlah_bayar_total = jumlah_bayar_jakarta + jumlah_bayar_yogyakarta + jumlah_bayar_bandung + jumlah_bayar_semarang + jumlah_bayar_surabaya
jumlah_bayar_total
```

```python
# Mengetahui rating tertinggi didominasi oleh tempat gratis atau berbayar
if jumlah_gratis_total > jumlah_bayar_total:
    print("Dikarenakan", jumlah_gratis_total, '>', jumlah_bayar_total)
    print("Sehingga didapatkan kesimpulan bahwa tempat dengan rating tertinggi di Indonesia didominasi oleh tempat yang gratis")
    
else:
    print("Dikarenakan", jumlah_gratis_total, '<', jumlah_bayar_total)
    print("Sehingga didapatkan kesimpulan bahwa tempat dengan rating tertinggi di Indonesia didominasi oleh tempat yang berbayar")
```

```python
## RDBMS dan DataFrame
# Import library pandas untuk membuat dataframes dan menginisiasi kolom-kolom yang akan diisi
import pandas as pd

column = ["Nama_Wisata", "Kategori", "Rating", "Nama_Tempat", "Kategori_Tempat", "Rating_Tempat"]

data = []

# Query SQL digunakan untuk menampilkan data yang ingin disimpan dalam dataframe. Row kemmudian ditambahkan pada data menggunakan fungsi append
import sqlite3

con = sqlite3.connect("Destinasi_Turis_di_Jakarta_dan_Surabaya.db", timeout=10)
cur = con.cursor()

rows = cur.execute("SELECT a.Nama_Wisata, a.Kategori, a.Rating, p.Nama_Tempat, p.Kategori_Tempat, p.Rating_Tempat FROM Jakarta as a JOIN Surabaya as p ON a.Id_Wisata=p.Id_Wisata")

for row in rows:
    data.append(row)
    
con.commit()

con.close()

# Menggunakan fungsi DataFrame pada pandas
df = pd.DataFrame(data, columns=column)
df
```

```python
# Rata-Rata Harga di Setiap Kota
import seaborn as sns
import matplotlib.pyplot as plt

harga_jakarta = df[df["Kota"] == "Jakarta"]["Harga"].mean()
harga_yogyakarta = df[df["Kota"] == "Yogyakarta"]["Harga"].mean()
harga_bandung = df[df["Kota"] == "Bandung"]["Harga"].mean()
harga_semarang = df[df["Kota"] == "Semarang"]["Harga"].mean()
harga_surabaya = df[df["Kota"] == "Surabaya"]["Harga"].mean()
harga_rata_rata = [harga_jakarta, harga_yogyakarta, harga_bandung, harga_semarang, harga_surabaya]

plt.figure(figsize=(7, 8))
sns.barplot(x=["Jakarta", "Yogyakarta", "Bandung", "Semarang", "Surabaya"], y=harga_rata_rata, palette='cubehelix')

# Loop melalui setiap bar dan nilainya
for i, value in enumerate(harga_rata_rata):
    # Dapatkan posisi x dan y dari bar
    x = i + 0.1  # Posisi x bar di tengah
    y = value + 0.1  # Posisi y sedikit di atas bar

    # Tambahkan label teks
    plt.text(x, y, f"{int(value)}", ha='center', va='bottom', fontsize=12)


plt.xlabel('KOTA')
plt.ylabel('HARGA')
plt.title('RATA RATA HARGA PADA TIAP KOTA')
                                  
plt.xticks(rotation=90)
plt.show()
```
